```python
import pickle
from matplotlib import pyplot as plt
import numpy as np
```

```python
with open("coldstart_source_train.pkl", "rb") as f:
    src_data = pickle.load(f)
with open("coldstart_target_train_36.pkl", "rb") as f:
    tgt_data = pickle.load(f)


fig = plt.figure(figsize=(12, 4))
ax1, ax2 = fig.add_subplot(1, 2, 1), fig.add_subplot(1, 2, 2)
ax1.plot(src_data[0], label="source domain")
ax1.plot(tgt_data[0], label="target domain")
ax1.set_xlabel("historical length")
ax1.tick_params(labelleft=False, left=False)
ax1.legend()
ax2.bar(0, len(src_data), label="source domain")
ax2.bar(1, len(tgt_data), label="target domain")
ax2.set_ylabel("number of data")
ax2.tick_params(labelbottom=False, bottom=False)
ax2.legend()
plt.show()
```

```python
with open("fewshot_source_train.pkl", "rb") as f:
    src_data = pickle.load(f)
with open("fewshot_target_train_20.pkl", "rb") as f:
    tgt_data = pickle.load(f)

fig = plt.figure(figsize=(12, 4))
ax1, ax2 = fig.add_subplot(1, 2, 1), fig.add_subplot(1, 2, 2)
ax1.plot(src_data[0], label="source domain")
ax1.plot(tgt_data[0], label="target domain")
ax1.set_xlabel("historical length")
ax1.tick_params(labelleft=False, left=False)
ax1.legend()
ax2.bar(0, np.log(len(src_data)), label="source domain")
ax2.bar(1, np.log(len(tgt_data)), label="target domain")
ax2.set_ylabel("log number of data")
ax2.tick_params(labelbottom=False, bottom=False)
ax2.legend()
plt.show()
```
