```python
import pickle
from matplotlib import pyplot as plt
import numpy as np
```

```python
with open("synthetic/coldstart_source.pkl", "rb") as f:
    src_data = pickle.load(f)
with open("synthetic/coldstart_target_valid_36.pkl", "rb") as f:
    tgt_data = pickle.load(f)


fig = plt.figure(figsize=(15, 5))
ax1, ax2 = fig.add_subplot(1, 2, 1), fig.add_subplot(1, 2, 2)
ax1.bar(0, len(src_data), label="source domain")
ax1.bar(1, len(tgt_data) * 5, label="target domain")
ax1.set_ylabel("number of data")
ax1.tick_params(labelbottom=False, bottom=False)
ax1.legend()
ax2.plot(src_data[0], label="source domain")
ax2.plot(tgt_data[0], label="target domain")
ax2.set_xlabel("historical length")
ax2.tick_params(labelleft=False, left=False)
ax2.legend()
plt.show()
```

```python
with open("synthetic/fewshot_source.pkl", "rb") as f:
    src_data = pickle.load(f)
with open("synthetic/fewshot_target_valid_20.pkl", "rb") as f:
    tgt_data = pickle.load(f)

fig = plt.figure(figsize=(15, 5))
ax1, ax2 = fig.add_subplot(1, 2, 1), fig.add_subplot(1, 2, 2)
ax1.bar(0, np.log(len(src_data)), label="source domain")
ax1.bar(1, np.log(len(tgt_data) * 5), label="target domain")
ax1.set_ylabel("log number of data")
ax1.tick_params(labelbottom=False, bottom=False)
ax1.legend()
ax2.plot(src_data[0], label="source domain")
ax2.plot(tgt_data[0], label="target domain")
ax2.set_xlabel("historical length")
ax2.tick_params(labelleft=False, left=False)
ax2.legend()
plt.show()
```
